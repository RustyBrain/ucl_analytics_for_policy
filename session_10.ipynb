```r
library(tidyverse)
```

```r
df <- read_csv('lab_data/ghw.csv')
```

```r
ggplot(data = df, mapping=aes(x=Height, y=Weight)) + geom_point() + geom_smooth(method="lm")
```

```r
model <- lm(Weight ~ Height, data=df)
```

```r
model
```

```r
predict(model, data.frame(Height=158), interval = "confidence")
```

```r
ggplot(data=df, aes(model$residuals)) + geom_histogram(binwidth = 1)
```

```r
summary(model)
```

```r
summary(model)$r.squared
```
